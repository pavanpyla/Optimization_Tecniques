```python
def f(X):
    return 5*X[0]+4*X[1]
def g1(x):
    return (24-6*x)/4
def g2(x):                                          #objective function and its subjective functions
    return (6-x)/2
def g3(x):
    return 1+x
def g4(x):
    return 2

# def find_int(L1,L2):
#     xi = (L1[1] - L2[1]) / (L1[0] - L2[0])
#     yi = L1[0] * xi + L1[1]
#     return xi,yi

def find_corner():
    """
    this functions takes no orguments ,using the coefficients of the functions ,it returns the Array of tuples of intersecting 
    points of given lines
    where X is the Array of the tuples
    """
    corner12 = np.array([[6, 4], [1, 2]])
    Cornerx = np.array([24, 6])
    X6 = np.linalg.inv(corner12).dot(Cornerx)
                                                                   

    corner34 = np.array([[0, 1], [-1, 1]])
    Cornerx = np.array([2, 1])
    X2 = np.linalg.inv(corner34).dot(Cornerx)
    X2

    corner24 = np.array([[0, 1], [1, 2]])
    Cornerx = np.array([2, 6])
    X3 = np.linalg.inv(corner24).dot(Cornerx)
    X3

    corner36 = np.array([[1, 0], [-1, 1]])
    Cornerx = np.array([0, 1])
    X4 = np.linalg.inv(corner36).dot(Cornerx)
    X4

    corner15 = np.array([[6, 4], [0, 1]])
    Cornerx = np.array([24, 0])
    X1 = np.linalg.inv(corner15).dot(Cornerx)
    X1

    corner56 = np.array([[0, 1], [1, 0]])
    Cornerx = np.array([0, 0])
    X5 = np.linalg.inv(corner56).dot(Cornerx)
    X5

    X = np.array([X1,X2,X3,X4,X5,X6])
    return X
```
