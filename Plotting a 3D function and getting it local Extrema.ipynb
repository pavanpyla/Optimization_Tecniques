## Ploting the given function using matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import minimize

f = lambda x1, x2: 4*x1+6*x2-2*x1*x1-2*x1*x2-2*x2*x2
x2 = np.linspace(-50,50,500)
x1 = np.linspace(-50,50,500)           #getting the poins

X, Y = np.meshgrid(x1,x2)
F = f(X,Y)


fig = plt.figure(figsize=(14,8))
ax = plt.subplot(projection='3d') 
                                    #setting the #D projection axis 


surface  = ax.plot_surface(X,Y,F, cmap= 'hot')
ax.scatter3D(1,-1,0,color="w")
fig.colorbar(surface , aspect=25)
ax.contour(X, Y, F, cmap ='hot')
ax.set_xlabel('x1', labelpad=10)
ax.set_ylabel('x2', labelpad=30)
ax.set_zlabel('f(X)', labelpad=10)




```

## Applying the Gradient Descent Method 

```python
import numpy as np

def f(T):
   
     return -1*(4*T[0]+6*T[1]-2*(T[0])**2-2*T[0]*T[1]-2*(T[1])**2)         #we took -ve of function so that we we can find minimum

def gradf(T):
   
    return np.array([  -4+4*T[0]+2*T[1]   , -6+2*T[0]+4*T[1]]  )             # Gradient of the given function

input1 = float(input("Inital Point 1 (x1) :"))
input2 = float(input(" Initial Point 2 (x2) :"))

T = np.array([input1,input2])                   #Initial Point


itrations = 1000
n = 1000
count=0

x1 = np.array([0,0])
err = np.array([0,0])

while(itrations):
    
    if(not np.allclose(gradf(T), err, rtol=0.001, atol=0.001, equal_nan=False)):   
        
        
        """Returns True if two arrays are element-wise equal within a tolerance.

        The tolerance values are positive, typically very small numbers.  The
        relative difference (`rtol` * abs(`b`)) and the absolute difference
        `atol` are added together to compare against the absolute difference
            between `a` and `b`.  """

        
        count=count+1
        print("Iteration =",count)
        print(f"the next point =   {T}")
        print(f"Gradient of function : {gradf(T)}\n")
        x1 = T - 0.01*gradf(T)
        
        T = x1
        itrations = itrations-1
        
    else:
        break

        





print(f"\n  The function is maximum at ={T} and the value is f(X) = ",-1*f(T))
```
